```python
df=pd.read_csv('/content/Iris (2).csv')
df.head()
```

```python
sns.scatterplot(data=df,x=df["SepalLengthCm"],y=df["SepalWidthCm"],color='black')
sns.scatterplot(data=df,x=df["PetalLengthCm"],y=df["PetalWidthCm"],color='orange',markers='.')
```

**Above the scatter Plot visualization Some parts are outlier **

```python
x=df.iloc[:,1:4].values
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

According to this elbow method Visualization the least of steepness of curve belong to 3 so we have 3 clusters identification

```python

plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

predict the optimum number of clusters
and represent it visually.

Optimum number of cluster is 3 according to above given diagram

```python
kmean=KMeans(n_clusters=3)
yhat=kmean.fit_predict(x)
center=np.array(kmean.cluster_centers_)
```

```python
plt.scatter(x[yhat==0,0],x[yhat==0,1],label="Iris setosa")
plt.scatter(x[yhat==1,0],x[yhat==1,1],label="Iris Versicolor")
plt.scatter(x[yhat==2,0],x[yhat==2,1],label="Iris Verginica")
plt.scatter(center[:,0],center[:,1],color='black',marker='D',label="Centroid")
plt.legend()
```
